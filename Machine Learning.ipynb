```python
missing_values = ["na","Na","NA","n/a","N/a","N/A","nan","Nan",""," ", np.nan]
df = pd.read_csv("obesity_data.csv", na_values = missing_values)
df
```

```python
df.isnull().any()
```

```python
sns.heatmap(df.isnull(), yticklabels=False)
```

```python
df["Gender"].replace("Male", "Alpha Male", inplace=True)
df
```

```python
df = df.fillna(df.mean())
df
```

```python
df.corr()
```

```python
df.describe()
```

```python
df['ObesityCategory'].ffill(inplace=True)
```

```python
df.replace("Normal weight", 0, inplace=True)
df.replace("Overweight", 1, inplace=True)
df.replace("Underweight", -1, inplace=True)
df.replace("Obese", 2, inplace=True)
```

```python
df
```

```python
X = df.drop(["Gender"],axis="columns")
Y = df["Gender"]
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.30)
```

```python
# apply svm
from sklearn import svm

clf = svm.SVC(kernel='rbf')
clf.fit(X_train, y_train)
y_pred = clf.predict(X_test)

# check accuracy
from sklearn.metrics import accuracy_score

accuracy = accuracy_score(y_test, y_pred)
print('Accuracy:', accuracy)


# parameter tuning
from sklearn.model_selection import GridSearchCV 
  
# defining parameter range 
param_grid = {'C': [0.1, 1, 10, 100, 1000],  
              'gamma': [1, 0.1, 0.01, 0.001, 0.0001], 
              'kernel': ['rbf']}  
  
grid = GridSearchCV(clf, param_grid, refit = True, verbose = 3) 
  
# fitting the model for grid search 
grid.fit(X_train, y_train) 

# print how our model looks after hyper-parameter tuning 
print(grid.best_estimator_) 
```
